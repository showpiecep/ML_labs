Опишите, что изменилось и почему: данные стандартизивированы => для каждой гистограммы наибольшее значение стало примерно около 0, также изменились диапазоны. Это нужно для того, чтобы легче обробатывала входные значения.

```python
import numpy as np

def mean(xs): 
    '''Среднее значение числового ряда'''
    return sum(xs) / len(xs)

def variance(xs):
    '''Дисперсия (варианс) числового ряда,
       несмещенная дисперсия при n <= 30'''
    mu = mean(xs)
    n = len(xs)
    n = n-1 if n in range(1, 30) else n  
    square_deviation = lambda x : (x - mu) ** 2 
    return sum( map(square_deviation, xs) ) / n

def standard_deviation(xs):
    '''Стандартное отклонение числового ряда'''
    return np.sqrt( variance(xs) )
```

```python
print('до стандартизации'.upper())
print(f"M(age) - {mean(df['age'])}, σ - {standard_deviation(df['age'])}")
print(f"M(creatinine_phosphokinase) - {mean(df['creatinine_phosphokinase'])}, σ - {standard_deviation(df['creatinine_phosphokinase'])}")
print(f"M(ejection_fraction) - {mean(df['ejection_fraction'])}, σ - {standard_deviation(df['ejection_fraction'])}")
print(f"M(platelets) - {mean(df['platelets'])}, σ - {standard_deviation(df['platelets'])}")
print(f"M(serum_creatinine) - {mean(df['serum_creatinine'])}, σ - {standard_deviation(df['serum_creatinine'])}")
print(f"M(serum_sodium) - {mean(df['serum_sodium'])}, σ - {standard_deviation(df['serum_sodium'])}\n")
print('после стандартизации'.upper())
print(f"M(age) - {mean(data_scaled[:,0])}, σ - {standard_deviation(data_scaled[:,0])}")
print(f"M(creatinine_phosphokinase) - {mean(data_scaled[:,1])}, σ - {standard_deviation(data_scaled[:,1])}")
print(f"M(ejection_fraction) - {mean(data_scaled[:,2])}, σ - {standard_deviation(data_scaled[:,2])}")
print(f"M(platelets) - {mean(data_scaled[:,3])}, σ - {standard_deviation(data_scaled[:,3])}")
print(f"M(serum_creatinine) - {mean(data_scaled[:,4])}, σ - {standard_deviation(data_scaled[:,4])}")
print(f"M(serum_sodium) - {mean(data_scaled[:,5])}, σ - {standard_deviation(data_scaled[:,5])}\n")
```

Стандартизированное значение вычисляется по формуле:

(x(i) - M(i)) / σ

```python
def from_old_range2new_range(old_value,
                             old_min=0.0,
                             old_max=1.0,
                             new_min=-5,
                             new_max=10):
    old_range = old_max - old_min  
    new_range = new_max - new_min  
    converted_value = ((old_value - old_min) * new_range / old_range) + new_min
    return converted_value
```
